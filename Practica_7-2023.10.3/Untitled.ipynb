```python
class SaldoInsuficienteError(Exception):
    pass

class CuentaBancaria:
    def __init__(self, saldo_inicial):
        self.saldo = saldo_inicial

    def retirar(self, cantidad):
        try:
            if cantidad < 0:
                raise ValueError("La cantidad a retirar no puede ser negativa")
            if cantidad > self.saldo:
                raise SaldoInsuficienteError("Saldo insuficiente en la cuenta")
            
            self.saldo -= cantidad
            return cantidad
        except ValueError as e:
            return str(e)
        except SaldoInsuficienteError as e:
            return str(e)

# Ejemplo de uso:
cuenta = CuentaBancaria(1000)

try:
    retiro = cuenta.retirar(500)
    print(f"Se retiraron {retiro} de la cuenta. Saldo actual: {cuenta.saldo}")
except Exception as e:
    print(f"Error: {e}")
```
